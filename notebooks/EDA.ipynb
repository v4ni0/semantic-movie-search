We will keep the concerts and others in the data, but we must see the different genres.

```python
genres = data['genres'].str.split('-').explode().value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=genres.values, y=genres.index, palette='Set1')
```

```python
genre_na_counts = data['genres'].isna().sum()
plt.pie([genre_na_counts, len(data) - genre_na_counts], labels=['NA Genres', 'Non-NA Genres'], autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
```

We see a really low na count, so let us remove it

```python
data.dropna(subset=['genres'], inplace=True)
```

```python
data
```

It would be interesting to see the average rating by genre

```python
avg_rating_per_genre = {}
for genre in genres.index:
    avg_rating_per_genre[genre] = data[data['genres'].str.contains(genre, na=False)]['vote_average'].mean()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=list(avg_rating_per_genre.keys()),
            y=list(avg_rating_per_genre.values()),
            palette='Set3')
plt.xticks(rotation=45)
```

Very good distribution, it looks like the data does not have much invalid or irrelevant rows left.

We should see again missing values

```python
missing_values= {}
for feature in data.columns:
    missing_values[feature] = data[feature].isna().sum()
plt.figure(figsize=(10, 6))
sns.barplot(x=list(missing_values.keys()), y=list(missing_values.values()), palette='Set2')
plt.xticks(rotation=45)
```

We will remove the credits and poster_path rows with missing values as they form a small percentage of all rows. We will also remove the status column as we are only using released movies

```python
data.drop(columns = ['status'], inplace=True)
```
